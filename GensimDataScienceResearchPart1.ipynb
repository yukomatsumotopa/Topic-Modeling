```python
# List a set of frequent words to eliminate them. 
stoplist = set('for a of the and in is to or be on do are if i an . but not so you your my their ? you about , at'.split(' '))

# 1. Lowercase each words, 2. Split it by white space, and 3. Filter out stopwords
texts = [[word for word in document.lower().split() if word not in stoplist]
         for document in raw_corpus]

# Print cleaned words
texts
```

```python
# Count word frequencies to eliminate less frequent words. 
# And print words with the numbers of their frequencies. 

from collections import defaultdict
frequency = defaultdict(int)
for text in texts:
    for token in text:
        frequency[token] += 1
        
print(frequency)
```

```python
# Only keep words that appear more than once
# This process was needed because words appeard only once can be typos or unsusal words that does not have meanings. 

processed_corpus = [[token for token in text if frequency[token] > 1] for text in texts]
processed_corpus
```

```python
# from gensim.models import Word2Vec, KeyedVectors
# from gensim import models
# import pandas as pd
# import nltk


# Import corpora
from gensim import corpora
from collections import defaultdict

# Adding the processed corpus in dictionary
dictionary = corpora.Dictionary(processed_corpus)
'dictionary'.encode('utf-8')

# Save the Dictionary, which will be converted into vector.
dictionary.save('uf.dict')
print(dictionary)
```

```python
# Vector
# Print all words with their ID in the dictionary. 

print(dictionary.token2id)
```

```python
from gensim.corpora import MmCorpus
from gensim.test.utils import get_tmpfile

# Corpus Vectorization
bow_corpus = [dictionary.doc2bow(text) for text in processed_corpus]

# Store the corpus to be used to train models.
corpora.MmCorpus.serialize("uf.mm", bow_corpus)
bow_corpus
```
