```python
# Part II Transformation and Extracting Topics

# log
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
import tempfile
import os.path

# A temporary file was created and to be used to save a temporary dictionary, and corpus created in the program I. 
TEMP_FOLDER = tempfile.gettempdir()
print('Folder "{}" will be used to save temporary dictionary and corpus.'.format(TEMP_FOLDER))
```

```python
# Transformation Interface
from gensim import corpora, models, similarities
from gensim.models import Word2Vec, KeyedVectors
# import nltk

# The dictionary and the corpus created in the Program I was called to join the Part II program. 
if os.path.isfile(os.path.join(TEMP_FOLDER, 'C:\\Windows\\System32\\uf.dict')):  
    dictionary = corpora.Dictionary.load(os.path.join(TEMP_FOLDER, 'C:\\Windows\\System32\\uf.dict'))
    corpus = corpora.MmCorpus(os.path.join(TEMP_FOLDER,"C:\\Windows\\System32\\uf.mm"))
    print("The data set was generated from the Program I.")  
else:
    print("Please run the Program I to generate data set")
```

```python
# To check if the dictionary were correctly imported. 

print(dictionary[0])
print(dictionary[1])
print(dictionary[2])
```

```python
# Transform the corpus into TFIDF model
tfidf = models.TfidfModel(corpus)
#model = gensim.models.KeyedVectors.load_word2vec_format(corpus, binary=True)

corpus_tfidf = tfidf[corpus]
for doc in corpus_tfidf:
    print(doc)
```

```python
# Convert it into lsi model
lsi = models.LsiModel(corpus_tfidf, id2word=dictionary, num_topics= 2) 

corpus_lsi = lsi[corpus_tfidf]
```

```python
# Print the extracted topics
lsi.print_topics(2)
```

```python
# The actual output of LSI
for doc in corpus_lsi:
    print(doc)
```

```python
# Saved the LSI converted corpus for future research. 

lsi.save(os.path.join(TEMP_FOLDER, 'model.lsi'))
# lsi = models.LsiModel.load(os.path.join(TEMP_FOLDER, 'model.list'))
```
